### Rapprochement de la table de l'ERP et de la boutique en ligne

On réalise une double jointure pour fusionner les 3 tables : erp-liaison-web

### Détection d'éventuelles erreurs de saisie sur les prix des produits

#### Méthode 1 : par l'écart interquartile (IQR)

Source : https://hersanyagci.medium.com/detecting-and-handling-outliers-with-pandas-7adbfcd5cad8

On commence par faire une représentation graphique du prix. On choisit le boxplot car adapté à la détection d'outliers :

```python
sns.boxplot(x = erp_web['price'])
```

On affiche les statistiques de la variable pour avoir une autre vue que le bopxlot :

```python
erp_web[['price']].describe()
```

Calcul de l'IQR (interquartile range) :

```python
IQR = erp_web['price'].quantile(0.75) - erp_web['price'].quantile(0.25)
IQR
```

On fixe les limites au-delà desquelles on considère les valeurs comme aberrantes :

Limite inférieure = Q1 - 1,5 x IQR

```python
lim_inf = erp_web['price'].quantile(0.25) - 1.5 * IQR
lim_inf

# lim_inf < min donc inutile ici
```

Limite supérieure = Q3 + 1,5 x IQR

```python
lim_sup = erp_web['price'].quantile(0.75) + 1.5 * IQR
lim_sup
```

Affichage des outliers selon le critère de l'IQR :

```python
erp_web[erp_web['price'] >= lim_sup]
```

On trouve 37 outliers.

Représentation de la distribution après nettoyage :

```python
erp_web_new = erp_web[erp_web['price'] <= lim_sup]

sns.boxplot(x = erp_web_new['price'])
```

#### Méthode 2 : le Z score

Source : https://www.simplypsychology.org/z-score.html

On choisit de se placer à un seuil de 95%. Le Z score doit donc (en valeur absolue) être inférieur à 2.

On affiche les outliers :

```python
erp_web['Z score'] = abs(stats.zscore(erp_web['price']))
erp_web[erp_web['Z score'] > 2]
```

On trouve 34 outliers.

Représentation de la distribution après nettoyage :

```python
sns.boxplot(x = erp_web[erp_web['Z score'] <=2]['price'])
```

Mise en perspective des deux méthodes :

```python
erp_web['outliers_1'] = erp_web['price'] > lim_sup
erp_web['outliers_2'] = erp_web['Z score'] >= 2
```

```python
fig, axes = plt.subplots(1, 2)
sns.set(rc={"figure.figsize":(30, 5)})
sns.scatterplot(data = erp_web, x = 'product_id', y = 'price', hue = 'outliers_1', ax = axes[0])
sns.scatterplot(data = erp_web, x = 'product_id', y = 'price', hue='outliers_2', ax = axes[1])
plt.title("Représentation du prix des produits et mise en évidence des outliers par les méthodes 1 et 2")
plt.show()
```

On peut remarquer que les deux méthodes ont repéré des outliers qui sont sensiblement les mêmes, avec un critère légèrement plus strict du côté de l'IQR, qui en détecte 3 de plus.
